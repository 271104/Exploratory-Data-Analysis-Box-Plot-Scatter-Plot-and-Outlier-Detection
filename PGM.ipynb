```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd
```

```python
titanic = sns.load_dataset('titanic') 
titanic['age'].fillna(titanic['age'].median(), inplace=True)
```

```python
# Box Plot 
plt.figure(figsize=(12, 5)) 
plt.subplot(1, 2, 1) 
plt.boxplot(titanic['age']) 
plt.title('Box Plot - Age') 
```

```python
plt.subplot(1, 2, 2) 
plt.boxplot(titanic['fare']) 
plt.title('Box Plot - Fare') 
plt.show() 
```

```python
# Outlier removal using IQR 
Q1_age = np.percentile(titanic['age'], 25) 
Q3_age = np.percentile(titanic['age'], 75) 
IQR_age = Q3_age - Q1_age 
titanic_cleaned = titanic[(titanic['age'] >= Q1_age - 1.5 * IQR_age) & (titanic['age'] <= Q3_age + 1.5 * 
IQR_age)] 
Q1_fare = np.percentile(titanic['fare'], 25) 
Q3_fare = np.percentile(titanic['fare'], 75) 
IQR_fare = Q3_fare - Q1_fare 
titanic_cleaned = titanic_cleaned[(titanic_cleaned['fare'] >= Q1_fare - 1.5 * IQR_fare) & 
(titanic_cleaned['fare'] <= Q3_fare + 1.5 * IQR_fare)]
```

```python
# Scatter plot 
plt.scatter(titanic_cleaned['age'], titanic_cleaned['fare'], alpha=0.5, color='purple') 
plt.xlabel("Age") 
plt.ylabel("Fare") 
plt.title("Scatter Plot - Age vs Fare") 
plt.grid(True) 
plt.show()
```
